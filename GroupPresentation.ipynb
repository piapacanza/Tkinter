```python
import pandas as pd

df = pd.read_csv("./data/Family_Income_and_Expenditure.csv", low_memory=False)

data = {
    "Region": df["Region"],
    "Household Head Sex": df["Household Head Sex"],
    "Household Head Age": df["Household Head Age"],
    "Household Head Marital Status": df["Household Head Marital Status"],
    "Highest Attainment": df["Household Head Highest Grade Completed"],
    "Job or Business Indicator": df["Household Head Job or Business Indicator"],
    "Family members": df["Total Number of Family members"],
    "Family members employed": df["Total number of family members employed"],
    "House Age": df["House Age"],
    "Household Income": df["Total Household Income"],
    "Food Expenditure": df["Total Food Expenditure"],
    "Source of Income": df["Main Source of Income"],
    "Restaurant and hotels Expenditure": df["Restaurant and hotels Expenditure"],
    "Alcoholic Beverages Expenditure": df["Alcoholic Beverages Expenditure"],
    "Tobacco Expenditure": df["Tobacco Expenditure"],
    "Clothes Expenditure": df["Clothing, Footwear and Other Wear Expenditure"],
    "House Rental Value": df["Imputed House Rental Value"],
    "Medical Care Expenditure": df["Medical Care Expenditure"],
    "Transportation Expenditure": df["Transportation Expenditure"],
    "Communication Expenditure": df["Communication Expenditure"],
    "Education Expenditure": df["Education Expenditure"],
    "Miscellaneous Goods and Services Expenditure": df["Miscellaneous Goods and Services Expenditure"]

}

new_export = pd.DataFrame(data)
new_export.to_csv("./new_data/Family_Income_and_Expenditure.csv")
```
